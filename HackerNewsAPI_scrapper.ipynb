```python
import time
```

```python
story_time = {}
response = requests.get(url + 'newstories.json')
new_stories = response.json()
for story in new_stories:
    response = requests.get(url + f'item/{story}.json')
    if response.status_code == 200:
        created_time = response.json().get('time')
        print(datetime.fromtimestamp(created_time))
        story_time[story] = created_time
    time.sleep(1)
```

```python
story_time
```

```python
import pandas as pd 
times = []
for time in story_time.values():
    times.append(pd.to_datetime(time, unit='s'))

import numpy as np
times = np.array(times)
delta_seconds = list(map(lambda x : x.seconds, times[:-1] - times[1:]))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(delta_seconds)
plt.title("Time interval between each story")
plt.ylabel('# of posts')
plt.xlabel('delta between new story (in seconds)')
```
